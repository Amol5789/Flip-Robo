```python
url1=driver.find_elements_by_xpath("//nav[@class='yFHi8N']//a")
url1
page_url=[]
for i in url1:
    page_url.append(i.get_attribute("href"))
page_url
```

```python
#scrap comment
comments=[]
for i in page_url:
    driver.get(i)
    comment1=driver.find_elements_by_xpath("//div[@class='t-ZTKy']/div[1]/div[1]/div")
    for i in comment1:
        comments.append(i.text)
comments
```
